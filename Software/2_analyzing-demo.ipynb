# Convert to Magnet Coords and Plot Center Slices

```python
from postprocessing.utilities.convert_to_magnet_coords import convert_to_magnet_coordinate_frame
from postprocessing.visualizer.plot_slices import plot_center_slices
from postprocessing.utilities.csv_importer import csv_import

filename_rawdat      = '_demo_data/measurement_1/raw_dat.csv'
filename_magnetcoords   = convert_to_magnet_coordinate_frame(filename_rawdat, probe='lakeshore', magnet_type='V1.0') # relies on centerpoint.csv which is in the same directory as filename_rawdat. 

field, x_values, y_values, z_values = csv_import(filename_magnetcoords, position_b0=5)

plot_center_slices(field, x_values, y_values, z_values, title='Center Slices of Bz, in Magnet-Coordinates', unit='mT')
```

# Extrempoints

Taking a look at the extrem points before and after any longer measurement makes sense to detect some drifts etc.

```python
from postprocessing.visualizer.extrem_points import evaluate_extrempoints

extrema_before = '_demo_data/measurement_1/extrempoints_raw_dat_before.csv'
extrema_after = '_demo_data/measurement_1/extrempoints_raw_dat_after.csv'

evaluate_extrempoints(extrema_before, extrema_after, position_b0=3) # the demo raw data has the main component of the B0 in x. so the index of this coloum in the csv file is 3.
```
